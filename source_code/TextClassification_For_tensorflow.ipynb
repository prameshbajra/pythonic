```python
import tensorflow as tf
import numpy as np
import pandas as pd
from nltk import word_tokenize
from collections import defaultdict, Counter
```

```python
wine_dataframe = pd.read_csv("../input_data/wine-reviews/winemag-data_first150k.csv", index_col = 0)
counter_wine_variety = Counter(wine_dataframe["variety"].to_list())
top_10_counter_wine_variety = {i[0]: index for index, i in enumerate(counter_wine_variety.most_common(10))}
wine_dataframe = wine_dataframe[wine_dataframe["variety"].map(lambda x: x in top_10_counter_wine_variety)]

wine_dataframe
```
