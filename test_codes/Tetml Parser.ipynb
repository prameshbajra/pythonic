```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(open("../input_data/differential-equations.tetml"), "xml")
content = soup.find("Content")
content_childrens = [i for i in content.children if i != "\n"]
```

```python
len(content_childrens)
```

```python
def get_coords_main_child(child):
    result = []
    boxes = child.find_all("Box")
    for box in boxes:
        result.append({"coords": box.get("llx")+','+box.get("lly")+','+box.get("urx")+','+box.get("ury")})
        
    return result
```

```python
value = []
for child in content_childrens:
    value.extend(get_coords_main_child(child))
```

# Co ords for each main child (para, table what ever)

```python
value
```

```python
def get_coords_words(words):
    result = []
    for word in words:
        box = word.find("Box")
        result.append({"coords": "{}, {}, {}, {}".format(box.get("llx"), box.get("lly"), box.get("urx"), box.get("ury"))})
    return result
```

```python
value = []
for child in content_childrens:
    words = child.find_all("Word")
    value.extend(get_coords_words(words))
```

# Co ords for each word ...

```python
value
```

```python
def get_coords_glyph(main_parent):
    result = []
    glyphs = main_parent.find_all("Glyph")
    for glyph in glyphs:
        result.append({"coords": "{}, {}".format(glyph.get("x"), glyph.get("y"))})
        
    return result
```

```python
value = []
for child in content_childrens:
    value.extend(get_coords_glyph(child))
```
